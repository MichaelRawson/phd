```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

length = 1024
frequencies = [16, 64, 256]

data = pd.DataFrame({})
activations = np.arange(0, length)
data["constant"] = np.array([1] * length)

for frequency in frequencies:
    data[f"decay:{frequency}"] = np.exp2(-np.floor(activations / frequency))

plt.figure(figsize=(6, 4))
sns.lineplot(data=data)
plt.xlabel("Activations")
plt.ylabel("AWR, proportional")
plt.savefig('shape-decay.pdf')
```

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

length = 1024
frequencies = [16, 64, 256]

data = pd.DataFrame({})
activations = np.arange(0, length)
data["constant"] = np.array([1] * length)

for frequency in frequencies:
    data[f"converge:{frequency}"] = 1 - np.exp2(-np.floor(activations / frequency))

plt.figure(figsize=(6, 4))
sns.lineplot(data=data)
plt.xlabel("Activations")
plt.ylabel("AWR, proportional")
plt.savefig('shape-converge.pdf')
```
